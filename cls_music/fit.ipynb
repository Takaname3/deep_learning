```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import pathlib
import warnings
from fastai.vision import *
warnings.filterwarnings('ignore')
```

```python
# transformation des musique en image
# on utilise des fichiers .wav car en local on a rencontr" des problemes avec librosa et les fichiers mp3
def audio2image(folder, destination):
    directory = pathlib.Path(folder)
    
    for filename in directory.glob('**/*.wav'):
        print(filename)      
        try:
            x, sample_rate = librosa.load(filename, offset=30,duration=30)

            n_fft = 1024
            hop_length = 256
            n_mels = 40
            fmin = 20
            fmax = sample_rate / 2 

            mel_spec_power = librosa.feature.melspectrogram(x, sr=sample_rate, n_fft=n_fft, 
                                                            hop_length=hop_length, 
                                                            n_mels=n_mels, power=2.0, 
                                                            fmin=fmin, fmax=fmax)
            mel_spec_db = librosa.power_to_db(mel_spec_power, ref=np.max)
            plt.imsave(pathlib.Path(destination).joinpath(filename.name[:-4]+".png"), mel_spec_db)
        except:
            print("Oops! Try again...")
        
```

```python
# 4 catégories de musiques différentes : classic, jazz, metal, et nightcore.
# Pour toutes les catégories : environs 200 images (apres transformation)

classes = ['classic','jazz','metal','nightcore']
for c in classes:
    audio2image("audio/sample/" + c, "img_data/" + c)
```

```python
# chargement des imgages + repartition 75% - 25% pour train - test
data = ImageDataBunch.from_folder(path="",train="train", test="test", valid_pct=0.25, ds_tfms=get_transforms(), bs=32).normalize(imagenet_stats)
```

```python
# utilisation du model resnet50, apres different essaie c'est qui donne le meilleur resultat
learn = cnn_learner(data, models.resnet50, metrics=accuracy)
```

```python
# recherche du lr
learn.lr_find()
```

```python
learn.recorder.plot(suggestion=True)
lr = learn.recorder.min_grad_lr
# celui ci etant plutot petit qu habituellement (ordre du 10^-4), il faudra peu plus de cycle d entrainement (environ 5-6)
```

```python
# on commence avec quatre epochs
learn.fit_one_cycle(4,slice(lr))
```

```python
# le train loss etant tres superieur au valid loss, on refait quelques epochs
learn.fit_one_cycle(2,slice(lr))
```

```python
# on save afin de faire du fit tuning
learn.save("etape1")
```

```python
# on relache les poids (unfreeze) et on refait la recherche des parametres
learn.unfreeze()
learn.lr_find()
learn.recorder.plot(suggestion=True)
```

```python
# on obtient le resultat 2.29E-06 avec lequel on refait quelques cycles
learn.fit_one_cycle(2,max_lr=slice(2.29e-06,lr/10))
learn.save("etape2")
```

```python
# apres plusieurs test pour aller plus loin, on y perd de plus en plus donc on s'arrete ici
# resultat de 89.2%, resultat plutot correct pour l'echantillonnage donné
```

```python
# predictions
preds,y,losses = learn.get_preds(with_loss=True)
```

```python
# recherche des plus grandes confusions
interp = ClassificationInterpretation(learn, preds, y, losses)
interp.plot_confusion_matrix(figsize=(8,8),dpi=60)
```

```python
interp.most_confused(3)
```
